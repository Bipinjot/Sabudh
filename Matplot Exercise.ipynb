```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig= plt.figure()
axes1= fig.add_axes([0,0,1,1])
axes2= fig.add_axes([0.2,0.5,.2,.2])

axes1.plot(x,y)
axes2.plot(x,y)
```

```python
fig= plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim(20,22)
axes2.set_ylim(30,50)
axes2.set_title('zoom')
```

```python
fig, axes= plt.subplots(1,2)
```

```python
axes[0].plot(x,y, lw=4, ls='--')
axes[1].plot(x,z,'r',lw=5)
```
