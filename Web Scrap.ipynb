```python
i=0
for n in range(1,5):
    print(n)
    a=requests.get('https://xkcd.com/'+str(n))
    soup=bs(a.content)
    images=soup.find_all('img')
    print(images)
    for i,img in enumerate(images):
        print(img['src'])
        url='https:'+img['src']
        print(i)
        print(url)
        try:
            urllib.request.urlretrieve(url,'Desktop/Sabudh/images/'+str(i)+str(n)+'.png')
        except Exception as err:
            print(err)
   
```
