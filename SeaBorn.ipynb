```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips= sns.load_dataset('tips')
flights= sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

```python
flights.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
sns.jointplot(x=tips['total_bill'],y='tip',data=tips, kind='reg')
```

```python
sns.pairplot(tips, hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips, kind='box')
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True, cmap='coolwarm')
```

```python
fc=flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
fc
```

```python
sns.heatmap(fc, cmap='magma',linewidths=.5)
```

```python
sns.clustermap(fc)
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(plt.scatter)
```

```python
g=sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time', 
           hue='day', aspect=1, size=4)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```
