```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('GRAPH')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
fig=plt.figure()
axes= fig.add_axes([0.1,0.1,0.0,0.1])
axes.plot(x,y)
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('BIG PLOT')

axes2.plot(y,x)
axes2.set_title('SMALL PLOT')
```

```python
fig= plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**3, label='CUBED')
ax.plot(x, x**2, label='SQUARED')
ax.legend()
```
